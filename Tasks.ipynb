# Задача 2
В зависимости от научной степени начисляется зарплата по формуле 
<br>
<br>
Аспирант (А) = 1 * количество лет опыта работы
<br>
<br>
Кандидат наук (B) = 2 * количество лет опыта работы
<br>
<br>
Доктор наук (C) = 3 * количество лет опыта работы
<br>
<br>
Пример: Для Кандидата Наук (B) c 5 годами опыта работы зарплата будет  2 * 5 = 10
<br>
<br>
Для Доктора Наук с 2 годами опыта работы зарплата будет  3 * 2 = 6

# Дано

```python
salary_multiplicator = {'A':1, 'B' : 2, 'C': 3}

employee_category = ['A','C','A', 'B']

employee_expirience = [1,2,8,3]

# Должно получится   [1, 6, 8, 6]
```

# Решение

```python
def salary_calculator(salar_mult, empl_cat, empl_exp):
    mapped_list = [salar_mult[k] for k in empl_cat] 
    salary_list = [a * b for a, b in zip(empl_exp, mapped_list)]
    
    return salary_list
```

```python
salary_calculator(salary_multiplicator, employee_category, employee_expirience)
```

Проверим на дополнительных данных

```python
salary_multiplicator_2 = {'A':5, 'B':1, 'C': 8}

employee_category_2 = ['C','C','B', 'A', 'A']

employee_expirience_2 = [8,3,8,4,2]

salary_calculator(salary_multiplicator_2, employee_category_2, employee_expirience_2)
```

# Задача 3
На столе стоят два стакана. В первом А литров воды, во втором А литров спирта.
<br>
Из стакана с водой взяли B литра воды, перелили в стакан со спиртом и тщательно перемешали.
<br>
Из второго стакана взяли B литра жидкости и перелили в первый стакан.
<br>
Какова концентрация воды в первом стакане?

Итак, я решал эту задачу поэтапно:
<br/><br/>
    1.  Из стакана с водой переливаем В литров. Концентрация воды в 1 стакане
<br/><br/> 
$$    
\frac{A-B}{A-B}
$$
<br>
    Концентрация воды во втором стакане
<br>
$$    
\frac{B}{A+B}
$$
<br>
    2. Берем из 2 стакана В литров жидкости (объем умножаем на концентрацию), то есть мы взяли воды объемом 
<br/><br/> 
$$    
B * \frac{B}{A+B}
$$
<br>
    3. Переливаем в первый стакан. Теперь концентрация воды в нём
<br/><br/> 
$$    
((А - B) + B * \frac{B}{A+B}) * \frac{1}{A}
$$
<br>
     Приведём к одному знаменателю 
<br/><br/> 
$$    
(\frac{(А - B)(А + В)}{A + В} + \frac{B * B}{A + B}) * \frac{1}{A}
$$
<br>
    Раскроем скобки
<br/><br/> 
$$    
\frac{A^2 - B^2 + B * B}{A + B} * \frac{1}{A}
$$
<br>
    Сокращаем дальше
<br/><br/> 
$$    
\frac{A^2}{A + B} * \frac{1}{A}
$$
<br>
    Сокращаем А, и вот финальная формула
<br/><br/> 
$$    
\frac{A}{A + B}
$$

# Задача 4
Чем вероятность отличается от максимального правдоподобия?

Цель метода максимального правдоподобия - найти оптимальные параметры распределения данных.
Допустим, у нас есть некие данные, и мы предполагаем, что они распределены каким то образом. Однако у нас нет параметров их распределения.
Для того, чтобы их найти, и используется метод максимального правдоподобия.

Правдоподобие выборки — это вероятность получения именно такой выборки, получения именно таких наблюдений / результатов,
т.е. произведение вероятностей получения каждого из результатов выборки.
Функция правдоподобия связывает правдоподобие выборки со значениями параметров распределения.

Откуда собственно идея – искать значение неизвестного параметра, при котором функция правдоподобия достигает максимума?
Истоки идеи проистекают из представления о том, что выборка – это единственный, доступный нам, источник знания о генеральной совокупности.
Все, что нам известно о генеральной совокупности, представлено в выборке. Поэтому, все, что мы можем сказать, так это то,
что выборка – это наиболее точное отражение генеральной совокупности, доступное нам.
Следовательно, нам требуется найти такой параметр, при котором имеющаяся выборка становится наиболее вероятной.

При вычислении вероятности, параметр истинной вероятности "зафиксирован". 
То есть, у нас есть оценка значения истинной вероятности, основываясь на которой, можно вычислить вероятность отдельных исходов.
Например, подкидывая монетку мы принимаем истинную вероятность за 0.5 и пытаемся оценить вероятность
каждого исхода при десяти испытаниях. На выходе получается функция вероятности.

При оценке правдоподобия зафиксированными оказываются данные. Например, нам может быть известен исход испытания. 
В этом случае мы можем попытаться вычислить, при каком значении параметра истинной вероятности
правдоподобие для этих конкретных данных примет максимальное значение.

Так же правдоподобие отличается от вероятностей в том, что оно не нормализовано в промежутке от 0 до 1.

# Задача 5
Почему в MSE используется 2 степень? А не 4, 6 и так далее?

MSE выводится с помощью функции правдоподобия. Давайте это и сделаем
    
Представим обыкновенную линейную регрессию. Допустим, решается задача нахождения цены дома.
Имеем вектор предикторов $X_1, X_2, X_3,... X_k$. - cостояние дома, расположение, количество спален и т.д.
Рассмотрим линейную модель 
$Y = X * \beta + \epsilon$ где X вектор, $\beta $ вектор.
А про ошибку $\epsilon$ предполагаем, что её мат ожидание = 0, дисперсия $\sigma^2$, а так же ошибка имеет нормальное распределение.
Предположим, что $\sigma^2$ известна и постоянна, а ошибки независимы и одинаково распределены.

$\epsilon = Y - X * \beta$
<br>
Плотность распределения
$\frac{1}{ \sqrt{2 \pi \sigma^2}} * exp \frac{-t^2}{2\sigma^2} $

Логарифм совместного распределения равен логарифму произведения за счет независимости 

$\sum_{i=1}^{n}log ( \frac{1}{ \sqrt{2 \pi \sigma^2}} * exp (- \frac{(y_i-X_i*\beta)^2}{2\sigma^2})) $ = 
$ - \frac{n}{2}log ( 2 \pi \sigma^2) - \sum_{i=1}^{n}\frac{(y_i-X_i*\beta)^2}{2\sigma^2}) $ 
<br>
Теперь нам нужно найти -$\beta$, при которых этот логарифм минимален

argmin ($ \frac{n}{2}log ( 2 \pi \sigma^2) + \sum_{i=1}^{n}\frac{(y_i-X_i*\beta)^2}{2\sigma^2}) $ )
<br>
argmin суммы, где одно из слогаемых - константа, позволяет нам эту константу откинуть. Или прибавить. А так же общий знаменатель выносим за скобку
<br>
argmin ($ \frac{1}{2\sigma^2}\sum_{i=1}^{n}(y_i-X_i*\beta)^2 $ )
<br>
Отбрасываем множитель перед суммой
<br>
argmin ($ \sum_{i=1}^{n}(y_i-X_i*\beta)^2 $ ) = 
<br>
argmin ($ \frac{1}{n}\sum_{i=1}^{n}(y_i-X_i*\beta)^2 $ ) = MSE
<br>
Какой здесь вывод? Если мы предполагаем, что цена дома имеет нормальное распределение, то метод максимального правдоподобия 
рекомендует использовать MSE. Мы получили некое теоретическое обоснование, что MSE действительно работает.
